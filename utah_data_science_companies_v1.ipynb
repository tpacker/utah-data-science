# Get Data

```python
wb = openpyxl.load_workbook(companies_filename, data_only=calc_xl_formulas)
print(type(wb))
print('sheet_names: ', wb.get_sheet_names())
```

```python
companies_sheet = wb.get_sheet_by_name('Companies')

companies_sheet
```

```python
companies_sheet.title
```

```python
data = companies_sheet.values
cols = next(data)[1:]
data = list(data)
idx = [r[0] for r in data]
data = (islice(r, 1, None) for r in data)
companies_df = DataFrame(data, index=idx, columns=cols)

companies_df.head(5)
```

```python
companies_df.sort_values('Score', ascending=False).head(7)
```
